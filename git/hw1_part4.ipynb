## Game of life with modified rules
we will run this part with julia language. Game of Life. Modified rules. 
0, 1, 2, 3 vote 0
4 vote does not change
5, 6, 7, 8 vote 1

```julia
using Random
using Plots
using Statistics

# function periodic_boundaries(i, j, N)
        
#         i > N ? 1 : i < 1 ? N : i
#         j > N ? 1 : j < 1 ? N : j
        
#     return CartesianIndex(i, j)
# end


function conway_gol(;N = 100, p = 0.5)
    n_cells = N^2
    initial_alive = floor(Int, p*n_cells)
    grid = zeros(Int, N, N)
    grid[randperm(n_cells)[1:initial_alive]] .= 1

    offsets = [CartesianIndex(i, j) for i in -1:1 for j in -1:1 if !(i == 0 && j == 0)]

    t_end = 200
    plt = heatmap(grid, aspect_ratio=1, title="p = $p", size=(300,300), color=["#134686", "#ED3F27"], colorbar=false,  clim=(0, 1),
        xticks=[],
        yticks=[],
        framestyle = :none) 
    savefig(plt, "conway0_p=$p.png")

    # anim = Animation()
    for t in 1:t_end
        grid_old = copy(grid)

        for i in 1:N, j in 1:N
            i_min = i == 1 ? N : i - 1
            i_max = i == N ? 1 : i + 1
            j_min = j == 1 ? N : j - 1
            j_max = j == N ? 1 : j + 1
            # j_min = max(j - 1, 1)
            # j_max = min(j + 1, N)

            # n_alive = sum(grid_old[i_min:i_max, j_min:j_max]) - grid_old[i, j]
            n_alive = grid_old[i_min, j_min] + grid_old[i, j_min] +  grid_old[i_max, j_min] + 
                    grid_old[i_min, j] + grid_old[i_max, j] + 
                    grid_old[i_min, j_max] + grid_old[i, j_max] + grid_old[i_max, j_max]

            if grid_old[i, j] == 1
                grid[i, j] = (n_alive ≤ 3) ? 0 : 1
            else
                grid[i, j] = (n_alive ≥ 5) ? 1 : 0
            end
        end

        plt = heatmap(grid, aspect_ratio=1, title="p = $p", size=(300,300), color=["#134686", "#ED3F27"], colorbar=false,  clim=(0, 1), 
        xticks=[],
        yticks=[],
        framestyle = :none)
        # frame(anim, plt)
    end
    
    savefig(plt, "conway_t=$t_end"*"_p=$p.pdf")
    # gif(anim, "conway_p=$p.gif", fps=20)
    
    return sum(grid)
end


function run()
    # ps = [0.3, 0.4, 0.45, 0.48, 0.50, 0.52, 0.55, 0.6, 0.7]
    ps = [0.45, 0.48, 0.50, 0.52, 0.55]
    # ps = [0.45, 0.55]
    n_tries = 10
    V = zeros(n_tries, length(ps))

    for t in 1:n_tries
    for (i,p) in enumerate(ps)
        vp = conway_gol(N=100, p=p)
        V[t, i] = vp
    end
    end
    

    # plt = plot(ps, mean(V, dims=1), error=std(V, dims=1), size=(400,400), xlabel="p", ylabel="V₁", markershape=:circle, linewidth=2, label="V₁(p)")  
    plt = plot(ps, mean(V, dims=1)[:], yerror=std(V, dims=1)[:],
            size=(400,400),
            xlabel="p", ylabel="V₁",
            linewidth=2,
            label="V₁(p)",
            framestyle=:box)
    savefig(plt, "conway_V_p.pdf")

end

@time run()
```
